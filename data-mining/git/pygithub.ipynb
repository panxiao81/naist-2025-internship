```python
spring_issues = g.search_issues(query='repo:spring-projects/spring-framework created:>2025-03-25 is:issue')
opensearch_issues = g.search_issues(query='repo:opensearch-project/OpenSearch created:>2025-03-25 is:issue')
lucene_issues = g.search_issues(query='repo:apache/lucene created:>2025-03-25 is:issue')
```

```python
def convert_issues_to_dataframe(issues): 
    data = dict.fromkeys(issues[0].raw_data.keys())
    for key in data.keys():
        data[key] = [d.raw_data[key] for d in issues]
    data['last_modified'] = [d.last_modified for d in issues]
    data['last_modified_datetime'] = [d.last_modified_datetime for d in issues]
    return pd.DataFrame(data)
```

```python
spring_df = convert_issues_to_dataframe(spring_issues)
spring_df
```

```python
spring_df.groupby([spring_df['last_modified_datetime'].dt.date]).size().plot()
plt.show()
```

```python
def convert_prs_to_dataframe(issues): 
    pr_list = [d.as_pull_request() for d in issues]
    data = dict.fromkeys(pr_list[0].raw_data.keys())
    for key in data.keys():
        data[key] = [d.raw_data[key] for d in pr_list]
    data['last_modified'] = [d.last_modified for d in pr_list]
    data['last_modified_datetime'] = [d.last_modified_datetime for d in pr_list]
    return pd.DataFrame(data)
```

```python
spring_pr_df = convert_prs_to_dataframe(spring_prs)
spring_pr_df
```

```python
workflow_runs = g.get_repo('spring-projects/spring-framework').get_workflow_runs()

def convert_workflow_runs_to_dataframe(issues): 
    data = dict.fromkeys(issues[0].raw_data.keys())
    for key in data.keys():
        data[key] = [d.raw_data[key] for d in issues[:20]]
    return pd.DataFrame(data)

workflow_df = convert_workflow_runs_to_dataframe(workflow_runs)
workflow_df
```
